ðŸ¼ Pandas â€“ 80% of Data Work

```python
import pandas as pd

# Load
df = pd.read_csv('data.csv')

# Inspect
df.head()
df.tail()
df.info()
df.describe()

# Select columns
df['amount']
df[['date', 'city']]

# Filter rows
df[df['amount'].notna()]
df[df['city'] == 'Delhi']
df[df['amount'] > 100]

# String ops
df[df['Investors Name'].str.contains('Sequoia', na=False)]

# Type conversion
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')

# Sort
df.sort_values('amount', ascending=False)
```

ðŸ§® GroupBy (When & How)

Use when: aggregation / summary needed

```python
# Count
df.groupby('city')['startup'].count()

# Sum
df.groupby('vertical')['amount'].sum()

# Multiple agg
df.groupby('city').agg({
    'startup': 'count',
    'amount': 'sum'
})

# Top N
(df.groupby('startup')['amount']
   .sum()
   .sort_values(ascending=False)
   .head(10))
```

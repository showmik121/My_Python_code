```python
points=pd.DataFrame(
    {
    '1st': [(3,4),(2,3),(2,2)],
    '2nd': [(1,2),(2,2),(3,3)]
    }
)
points
```

```python
def euclidian(row):
    pt_A=row['1st']
    pt_B=row['2nd']
    return (pt_A[0]-pt_B[1])
```

```python
points.apply(euclidian,axis=1)
```

concatenate

```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'id': [4, 5, 6],
    'name': ['D', 'E', 'F']
})

result = pd.concat([df1, df2])
print(result)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'id': [4, 5, 6],
    'name': ['D', 'E', 'F']
})

result = pd.concat([df1, df2],ignore_index=True)
print(result)
```
